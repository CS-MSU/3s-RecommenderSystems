```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv', index_col=['user_id', 'item_id'])
```

```python
train_labels = df['rating']
train_data = df.drop('rating', axis=1)
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model = CatBoostClassifier(
    task_type="GPU",
    devices='0',
    verbose=0
)
```

```python
model_params = {
    'depth': [4, 6, 8, 10, 12],
    'iterations': [200, 400, 600, 800, 1000],
    'learning_rate': [0.03, 0.1], 
    'l2_leaf_reg': [1, 3, 5, 7, 9]
}
```

```python
grid_search_result = model.grid_search(model_params, 
                                       X=train_data, 
                                       y=train_labels, 
                                       plot=False)
```

```python
grid_search_result # d 10, l2 9, it 100, lr 0.1
```

```python
model = CatBoostClassifier(
    iterations=1000,
    depth=10,
    learning_rate=0.1,
    l2_leaf_reg=9,
    task_type="GPU",
    devices='0',
    verbose=0
)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    train_data, train_labels, test_size=0.33, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(pred, y_test))
```
